```python
X=df['GPA'] # independent varialble
y=df['SAT']
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
X_train = X_train.to_numpy().reshape(-1, 1)
y_train = y_train.to_numpy().reshape(-1, 1)

X_test = X_test.to_numpy().reshape(-1, 1)
y_test=y_test.to_numpy().reshape(-1,1)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python


X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
plt.scatter(X_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
plt.scatter(X_train,y_train)
# for plotting the regression line
plt.plot(X_train,regressor.predict(X_train),color='red')
# for plotting the regression line using scatter plot
#plt.scatter(X_train,regressor.predict(X_train))
```

```python
y_pred=regressor.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
##  for checking the accuracy of the model  we use r2_score, mean_squared_error, mean_absolute_error,adjusted_r2_score

from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
mae=mean_absolute_error(y_test,y_pred)
mae
```

```python
score=r2_score(y_test,y_pred)
score
```

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
