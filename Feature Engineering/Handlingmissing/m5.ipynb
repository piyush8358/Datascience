```python
import numpy as np
list_marks=[45,31,27,50,29,33,40,38,42,35,39,48,32,41,49,39,31,26,38,33,40,41,36,28,45,50,48,39,40,4,1001,14]
```

```python
np.percentile(list_marks,[75])
```

```python
## [lower fence, upper fence]
Q1=np.percentile(list_marks,[25])
print(Q1)
```

```python
minimum,Q1,Q2,Q3,maximum=np.quantile(list_marks,[0,0.25,0.50,0.75,1.0])
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*IQR
higher_fence=Q3+1.5*IQR
```
