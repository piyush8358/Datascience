### Imbalanced Dataset Handling

* up-sampling
* down-sampling

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(123)
n_sample = 1000

class_0_ratio=0.9
n_class_0 = int(n_sample * class_0_ratio)
n_class_1 = n_sample - n_class_0
```

```python
n_class_0,n_class_1
```

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.tail()
```

```python
df['target'].value_counts()
```

### Up-sampling

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_majority.head()
```

```python
df_minority.head()
```

```python
# Upsample perform

from sklearn.utils import resample
```

```python
df_minority_upsmapled=resample(df_minority,
                               replace=True, # sample with replacement
                               n_samples=len(df_majority), # to match the majority class
                               random_state=42)
```

```python
df_minority_upsmapled.shape
```

```python
df_unsmapled=pd.concat([df_majority,df_minority_upsmapled])
```

```python
df_unsmapled['target'].value_counts()
```

### Down-sampling

```python
## CREATE MY DATAFRAME WITH IMBALANCED DATASET
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc=0, scale=1, size=n_class_0),
    'target': [0] * n_class_0
})

class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc=2, scale=1, size=n_class_1),
    'target': [1] * n_class_1
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```

```python
df.tail()
```

```python
df_minority=df[df['target']==1]
df_majority=df[df['target']==0]
```

```python
df_majority_unsampled=resample(df_majority,
                               replace=False, # sample with replacement
                               n_samples=len(df_minority), # to match with minority class
                                 random_state=42)
```

```python
df_majority_unsampled.shape
```

```python
df_unsmapled=pd.concat([df_majority_unsampled,df_minority])
```

```python
df_unsmapled['target'].value_counts()   
```
