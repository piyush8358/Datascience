```python
x=np.linspace(0,10,200)
```

```python
y=np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my plot")
plt.plot(x,y)
```

```python
y=np.random.rand(50)
x=np.random.rand(50)
plt.scatter(x,y)
```

```python
colur=np.random.rand(50)
sizes=1000*np.random.rand(50)
plt.scatter(x,y,c=colur,s=sizes,alpha=.2)
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my plot")
plt.plot(x,y)
```
